Definición de una LA ponderada. Se lee del nodo 0: hacia nodo 2 existe un peso de 4, del nodo 0: hacia nodo 7 existe un peso de 8 y del nodo 0: hacia nodo 14 existe un peso de 3....

```python
%%file 1.in
1|4 2|4
2|2 0|4
0|4 1|2 3|3 5|2 4|4
2|3 4|3
2|4 3|3
2|2 4|3
```

Funciones que permiten leer(readAdjl) y visualizar(adjlShow) una lista de adyacencia.

Visualizamos el grafo ponderado a partir de la lista de Adyacencia ponderada contenida en el archivo 1.in

```python
G, _ = readAdjl("1.in", weighted=True)
for i, edges in enumerate(G):
  print(f"{i:2}: {edges}")
adjlShow(G, weighted=True, layout="neato")
```

```python
path = kruskal(G)
adjlShow(G, weighted=True, layout="neato", path=path, simplepath=False)
```

```python
edges = [(0, 1, 4), (0, 2, 2), (1, 2, 4), (1, 3, 3), (1, 5, 2), (2, 3, 3), (2, 4, 4), (2, 5, 3), (3, 4, 3), (4, 5, 2)]
```

```python
FG = [[] for _ in range(6)]
for e in edges:
    u, v, w = e

    FG[u].append((v, w))

for i, edges in enumerate(FG):
  print(f"{i:2}: {edges}")
adjlShow(FG, weighted=True, layout="neato")
```

```python
path = kruskal(FG)
adjlShow(FG, weighted=True, layout="neato", path=path, simplepath=False)
```

Otra forma de crear el grafo y sus aristas ponderadas (sin leer una lista de adyacencia de sde un archivo).

```python
g = Graph(6)
g.addEdge(0, 1, 4)
g.addEdge(0, 2, 4)
g.addEdge(1, 2, 2)
g.addEdge(1, 0, 4)
g.addEdge(2, 0, 4)
g.addEdge(2, 1, 2)
g.addEdge(2, 3, 3)
g.addEdge(2, 5, 2)
g.addEdge(2, 4, 4)
g.addEdge(3, 2, 3)
g.addEdge(3, 4, 3)
g.addEdge(4, 2, 4)
g.addEdge(4, 3, 3)
g.addEdge(5, 2, 2)
g.addEdge(5, 4, 3)
```

Ejecutamos la funcion del Algoritmo Kruskal, **kruskal_algo()** a fin de obtener el costo total del mst para el grafo G.

```python
path = g.kruskal_algo()
```

```python
print(enumerate(path))
```

```python
adjlShow(G, weighted=True, layout="neato", path=path, simplepath=False)
```
