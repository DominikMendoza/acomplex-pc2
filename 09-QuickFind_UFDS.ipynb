# **Algoritmo Quick-Find**

Utilizamos la libreria graphviz y la funcion drawUF(lista de ids) para visualizar que nodos estan conectados (los que tienen el mismo id).

```python
import graphviz as gv
```

```python
def drawDS(ds):
  graph = gv.Digraph("DisjointSet")
  graph.graph_attr['rankdir'] = "BT"
  for e, p in enumerate(ds):
    graph.node(str(e))
    if p != e:
      graph.edge(str(e), str(p))
  return graph
```

```python
ds = [0, 0, 9, 9, 9, 6, 6, 7, 8, 9]
drawDS(ds)
```

#### **Find de componentes disjuntos**

Definimos la funcion **quickfind**, la cual recibe un arreglo de ids y un elemento a buscar en el (devuelve el id del item buscado)

```python
def quickfind(s, a):
    return s[a]
```

```python
#Obtenenos los id de cada uno de los items del conjunto
quickfind(ds, 1), quickfind(ds, 3), quickfind(ds, 4), quickfind(ds, 5), quickfind(ds, 7), quickfind(ds, 8)
```

Definimos la funcion qf_union(), que recibe un arreglo de ids y une los componentes a y b (asigna el id del elemento b a todos los elementos que tengan el mismo id del elemento a)

#### **Union de componentes disjuntos**

```python
def qf_union(s, a, b):
    pa = quickfind(s, a)
    pb = quickfind(s, b)
    for i in range(len(s)):
        if s[i] == pa:
            s[i] = pb
```

```python
ds = [i for i in range(10)]
print(ds)
drawDS(ds)
```

Unimos los elementos 3 y 4 del arreglo **s**.  La funcion qf_union() cambia los id del componente 3 por el id del componente 4.

```python
qf_union(ds, 3, 4)
print(ds)
drawDS(ds)

     
```
